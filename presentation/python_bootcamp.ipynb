# control structures and loops 
- Loops in Python are used to repeat a block of code multiple times until a certain condition is met. They help automate repetitive tasks, making your code more efficient and concise.

    - if else
    - for 
    - while 

```python
## if else
'''  
if True:
    run block

else:
    run block
''' 
x=20

if x%2==0:
    print(f'{x} is even number')
    if x%10==0:
        print(f'{x} is also divisible by 10')
    elif x%4==0:
        print(f'{x} is also divisible by 4')
    else:
        print(f'{x} is only divisible by 2')
else:
    print(f'{x} is odd')
```

```python
## for loop 
# Used to iterate over a sequence (e.g., list, tuple, string, range, etc.).
''' 
for item in list:
    block statements
'''

fruits=['apple','mango','banana','orange']
print(fruits)

# iterating over the loop 
# for fruit in fruits:
#     if fruit=='apple':
#         print('there is apple in the list')

for i in range(0,len(fruits)):
    print(fruits[i])
```

```python
# while loop 
# a while loop doesnt run for predefined number of iterations like the for loop. Instead, it stops as a soon as a given condition becomes true/false
''' 
while True:
    block statements
'''
num=1
while(num%10!=0):
    print(num)
    num=num+1
```

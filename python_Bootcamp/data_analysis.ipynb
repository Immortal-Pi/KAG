```python
# 1. Check for missing values in the datasets
# 2. Check the data types of all the columns 
laptop_data.isna().sum()
laptop_data.info()
```

```python
import re 

def extract_cpu_manufacturer(cpu):
    return cpu.split()[0]

def extract_cpu_series(cpu):
    return " ".join(cpu.split()[:2])

def extract_cpu_model(cpu):
    matches= re.search('\d+[A-Za-z]*',cpu)
    return matches.group() if matches else None 

def extract_clock_speed(cpu):
    matches=re.search('[\d\.]+GHz',cpu)
    return float(matches.group()[:-3]) if matches else None 


# laptop_data['Cpu'].apply(extract_cpu_manufacturer)
# laptop_data['Cpu'].apply(extract_cpu_series)
# laptop_data['Cpu'].apply(extract_cpu_model)
# laptop_data['Cpu'].apply(extract_clock_speed)
```

```python
laptop_data['CPU Manufacturer'] = laptop_data['Cpu'].apply(extract_cpu_manufacturer)
laptop_data['CPU Series'] = laptop_data['Cpu'].apply(extract_cpu_series)
laptop_data['CPU Model'] = laptop_data['Cpu'].apply(extract_cpu_model)
laptop_data['Clock Speed (GHz)'] = laptop_data['Cpu'].apply(extract_clock_speed)

laptop_data.head()
```

## task 3: Exploratory data Analysis (EDA) and Visualization 

```python
# box plot to check price distribution 
plt.boxplot(x=laptop_data['Price_euros'])
plt.title('Price Distribution')
plt.ylabel('Price')
plt.show()
```

```python
# Average price by brand
average_prices=laptop_data.groupby('Company')['Price_euros'].mean()
# average_prices.plot(kind='bar')
plt.bar(average_prices.index,average_prices.values)
plt.xlabel('Company')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.plot()
```

```python
# Average relation between CPU manufacturer and price 
cpu_manu=laptop_data.groupby('CPU Manufacturer')['Price_euros'].mean()
cpu_manu.plot(kind='bar')
```

```python
# identy the relation between GPU and price 
def gpu_price(gpu):
    return gpu.split()[0]

laptop_data['GPU_brand']=laptop_data['Gpu'].apply(gpu_price)
laptop_data.head()
```

```python
plt.bar(laptop_data['GPU_brand'],laptop_data['Price_euros'])
plt.ylabel('Price')
plt.show()
```

```python
# line plot
laptop_inches=laptop_data.groupby('Inches')['Price_euros'].mean() 
plt.xlabel('Inches')
plt.ylabel('Price')
plt.plot(laptop_inches.index,laptop_inches.values)
```
